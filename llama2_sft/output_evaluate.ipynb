```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('outputs/test_all_option_output_withGroundTruth.csv')
df
```

```python
c = 0
#label_list = []
indicator_list = []
for i in tqdm(range(len(df))):
    #print(i)
    ground_truth = df['ground_truth'][i].lower()
    output = df['option_1_ft_output'][i].lower()
    last_r_index = output.rfind('range')
    label = output[last_r_index:last_r_index+7]
    if ground_truth == label:
        c += 1
        indicator_list.append(1)
    else:
        indicator_list.append(0)
    #label_list.append(label)
    #print(label, ground_truth)

c/len(df)

df['option_1_ft_ind'] = indicator_list
```

```python
df.to_csv("outputs/test_all_option_output_withGroundTruth.csv", index = False)
```
